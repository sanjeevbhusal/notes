### the data set contains 2 data points
- a house with 1000 square feet sold for 300,0000
- a house with 2000 square feet sold for 500,000

The x-axis will represent square feet and y-axis will represent sold prize. The unit of representation will be in 1000's for both size and prize.

```python
import numpy as np
import matplotlib.pyplot as plt

# we represent input/feature variables with x
x_training = np.array([1.0, 2.0])
# we represent output/target variable with y
y_training = np.array([300.0, 500.0])
print(f"x_training data = {x_training}")
print(f"y_training_data = {y_training}")
```

```python
# we represent total training samples with m
m = x_training.shape[0]
print(f"Number of training examples is = {m}")
```

```python
# plot the datapoints
plt.scatter(x_training, y_training, marker='x', c='r')
# set the title
plt.title("Housing Prices")
# set the x-axis label
plt.xlabel("Size of house (1000s of sqft)")
# set the y-axis label
plt.ylabel("Price of house (1000s of dollar)")
plt.show()
```

 the model function for linear regression (which is a function that maps from `x` to `y`) is represented as 

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$

lets start with the value of $w=100$ and $b=100$

```python
w = 100
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

For both of the data points, lets calculate the value of the function.

- First data point is $x^{(0)}$ : f_wb = 200 * x[0] + 100
- Second data point is $x^{(1)}$: f_wb = 200 * x[1] + 100

ndarray (m,) defines a n-dimensional numpy array with the shape m 
(scalar) defines a argument with just maginitude and not dimensions.

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of the linear model
    Args:
        x (ndarray (m,)): Data, m examples
        w, b (scalar): Model parameters
    Returns
        y (ndarray (m, )): target values
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_training, w, b)
print(f"models prediction is = {tmp_f_wb}")
```

```python
# plot the models prediction
plt.plot(x_training, tmp_f_wb, c='b',label='Our Prediction')

# plot the data points
plt.scatter(x_training, y_training, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

## as we can see using w as 200 and b as 100 creates a line that perfectly fits our data. We can now use this function to predict the price for any house.
